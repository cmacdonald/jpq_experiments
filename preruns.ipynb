```python
INDEX_DIR = "/root/nfs/jpq/indices/"
INDEX_LOC = {
    "tct_colbert" : INDEX_DIR + "tct-hnp",
    "tas_b" : INDEX_DIR + "tasb",
    "star" : INDEX_DIR + "star",
    "repllama" : INDEX_DIR + "repllama7b",
    "e5": INDEX_DIR + "e5",
}
PYTHON = "/opt/miniconda3/envs/py312_jpq_dr/bin/python"
RUN = "/root/nfs/jpq/pyterrier_dr_jpq/pyterrier_dr/jpq/run.py"
EXP = "/root/nfs/jpq/experiments/"
```

# investigate PQ training sample size


```python
for model in ["tct_colbert"]:
    for M in [48, 96]:
        for PQSS in [10_000, 50_000, 159_744, 250_000]:
            for ibn in ["--in-batch-negs"]:
                args = [PYTHON, RUN]
                args += ["--model-name", model]
                args += ["--no-lambda-rank"]
                args += ["--base-index", INDEX_LOC[model]]
                args += ["--target-dir", EXP + "pqss"]
                args += ["--M", str(M)]
                args += ["--pq-sample-size", str(PQSS)]
                args += [ibn]
                args += [f" 2>&1 | tee {EXP}/pqss/m{M}-pqss{PQSS}.log"]
                print(" ".join(args))
```

## scan the M parameters for 3 models

```python

for model in ["tct_colbert", "star", "tas_b"]:
    for M in [24, 48, 96, 128]:
            for ibn in [("ibn", "--in-batch-negs"), ("no-ibn", "--no-in-batch")]:
                args = [PYTHON, RUN]
                args += ["--model-name", model]
                args += ["--no-lambda-rank"]
                args += ["--base-index", INDEX_LOC[model]]
                args += ["--target-dir", EXP + "Mscan"]
                args += ["--M", str(M)]
                args += [ibn[1]]
                args += [f" 2>&1 | tee {EXP}/Mscan/{model}_m{M}_{ibn[0]}.log"]
                print(" ".join(args))
```

```python
import os
for model in ["e5"]:
    for M in [96]:
        for nbits in [8]: #, 1 4, 6, 8, 10, 12
            for ibn in [("ibn", "--in-batch-negs")]: #, ("no-ibn", "--no-in-batch")]:
                LOG =  f"{EXP}/basesetting/{model}_m{M}_nbits{nbits}_{ibn[0]}_jpqneg200_lr.log"
                if os.path.exists(LOG):
                    continue
                args = [PYTHON, RUN]
                args += ["--model-name", model]
                args += ["--lambda-rank"]
                args += ["--jpq-negs", "200"]
                args += ["--base-index", INDEX_LOC[model]]
                args += ["--target-dir", EXP + "basesetting"]
                args += ["--M", str(M)]
                args += ["--nbits", str(nbits)]
                args += [ibn[1]]
                args += [f" 2>&1 | tee {LOG}"]
                print(" ".join(args))
```
