## ablation study on TCT

```python
DIRNAME="ablation_amd_3090"
model = "tct_colbert"
M=96
nbits=8
import os
for ibn in [("ibn", "--in-batch-negs"), ("no-ibn", "--no-in-batch")]:
    for qe in [("frozen", "--frozen-query-encoder"), ("unfrozen", "")]:
        for lr in [("lr", "--lambda-rank"), ("nolr", "")]:
            for jpqnegs in [("jpqneg200", ["--jpq-negs", "200"]), ("nojpqnegs", [])]:
                if qe[0] == "frozen":
                    mask = [ibn[0]=="ibn", lr[0]=="lr", jpqnegs[0]=="jpqneg200"]
                    if any(mask) and not all(mask):
                        continue
                if jpqnegs[0] == "nojpqnegs" and lr[0] == "lr":
                    continue
                LOG =  f"{EXP}/{DIRNAME}/{model}_m{M}_nbits{nbits}_{ibn[0]}_{jpqnegs[0]}_{lr[0]}_{qe[0]}.log"
                if os.path.exists(LOG):
                    continue
                args = ["(cat /proc/cpuinfo ; nvidia-smi; ", PYTHON, RUN]
                args += ["--model-name", model]
                args += [lr[1]]
                args += jpqnegs[1]
                args += ["--base-index", INDEX_LOC[model]]
                args += [qe[1]]
                args += ["--target-dir", EXP + "/"+ DIRNAME]
                args += ["--M", str(M)]
                args += ["--nbits", str(nbits)]
                args += [ibn[1]]
                args += [f" 2>&1) | tee {LOG}"]
                print(" ".join(args))
```

### PQ only!

```python
PYTHON_LION = "/opt/miniconda3/envs/py312_lion/bin/python"
DIRNAME="pq_only"
import os
for model in ["lion"]: #"repllama", "lion"  "tct_colbert", "e5", "tas_b", "star", "dragon"
        # "adore_star" # tct_colbert" #"e5", "tas_b", "star", "dragon"
    for M in [128,256]:
        for nbits in [8]: #, 1 4, 6, 8, 10, 12
            LOG =  f"{EXP}/{DIRNAME}/{model}_m{M}_nbits{nbits}.log"
            if os.path.exists(LOG):
                continue
            if model == "lion":
                args = [PYTHON_LION, RUN]
            else:
                args = [PYTHON, RUN]
            args += ["--model-name", model]
            args += ["--pq-only"]
            args += ["--base-index", INDEX_LOC[model]]
            args += ["--target-dir", f"{EXP}/{DIRNAME}"]
            args += ["--M", str(M)]
            args += ["--nbits", str(nbits)]
            args += [f" 2>&1 | tee {LOG}"]
            print(" ".join(args))
```
